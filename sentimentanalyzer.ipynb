```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df=pd.read_csv('train.txt',sep=';',header = None,names =['text','emotion'])
df.head()
```

```python
df.isnull().sum()
```

```python
df['emotion'].unique()
```

```python
unique_emotions=df['emotion'].unique()
emotion_number= {}
i=0
for emo in unique_emotions:
    emotion_number[emo]=i
    i+=1
df['emotion']= df['emotion'].map(emotion_number)
df.head()
```

```python
df['text']=df['text'].apply(lambda x : x.lower())
```

```python
import string
def remove_punc(txt):
  return txt.translate(str.maketrans('','',string.punctuation))

df['text']=df['text'].apply(remove_punc)
```

```python
def remove_numbers(txt):
  new=""
  for i in txt:
    if not i.isdigit():
      new=new+i
  return new

df['text']=df['text'].apply(remove_numbers)
```

```python
import re

def remove_urls(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r'', text)
df['text']=df['text'].apply(remove_urls)
```

```python
def remove_emojis(txt):
  new=""
  for i in txt:
    if i.isascii():
      new+=i
  return new

df['text']=df['text'].apply(remove_emojis)
```

```python
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
stopwords= set(stopwords.words('english'))
stopwords
```

```python
len(stopwords)
```

```python
nltk.download('punkt_tab')
```

```python
def remove(txt):
  words= word_tokenize(txt)
  cleaned =[]
  for word in words:
    if word not in stopwords:
      cleaned.append(word)
  return " ".join(cleaned)
```

```python
df['text']=df['text'].apply(remove)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df['text'],df['emotion'],test_size=0.20,random_state=42)
```

```python
X_test.value_counts()
```

```python
df.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer
```

```python
bow_vectorizer = CountVectorizer()
X_train_bow = bow_vectorizer.fit_transform(X_train)
X_train_bow
```

```python
X_test_bow = bow_vectorizer.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

```python
nb_model = MultinomialNB()
nb_model.fit(X_train_bow,y_train)
```

```python
pred_nb = nb_model.predict(X_test_bow)
```

```python
print(accuracy_score(y_test,pred_nb))
```

```python
tfidfv = TfidfVectorizer()
```

```python
X_train_tf = tfidfv.fit_transform(X_train)
X_test_tf = tfidfv.transform(X_test)

nb2_model = MultinomialNB()
nb2_model.fit(X_train_tf,y_train)
```

```python
pred_nb2 = nb2_model.predict(X_test_tf)
print(accuracy_score(y_test,pred_nb2))
```

```python
from sklearn.linear_model import LogisticRegression
logis_model = LogisticRegression(max_iter=1000)
logis_model.fit(X_train_tf,y_train)
pred_logis = logis_model.predict(X_test_tf)
print(accuracy_score(y_test,pred_logis))
```

```python
logis_model = LogisticRegression(max_iter=1000)
logis_model.fit(X_train_bow,y_train)
pred_logis = logis_model.predict(X_test_bow)
print(accuracy_score(y_test,pred_logis))
```

```python
import pickle

# Save the logis_model to a file
with open('logistic_regression_model.pkl', 'wb') as file:
    pickle.dump(logis_model, file)

print('Logistic Regression model saved successfully as logistic_regression_model.pkl')
```
